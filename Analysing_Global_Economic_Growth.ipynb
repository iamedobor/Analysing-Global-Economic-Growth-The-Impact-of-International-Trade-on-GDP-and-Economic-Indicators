**Check for outliers and handle them. This step is important in data analysis and model performance optimisation as it can significantly impact statistical metrics, influence measures of central tendency (mean, median and mode) and potentially lead to bias conclusions.**

**To check for outliers, we will employ the Z - Score model to measure how many standard deviations a data point is from the mean of a distribution. This will provide a standardised way to assess the relative position of a value within a dataset. Note that positive Z-scores will indicate values above the mean, while negative Z-scores will indicate values below the mean.**

```python
# Import Library
import numpy as np

# Note that we are not using all the variables in our dataset. We are using a few that is related to our analytical question.

# Selecting specific columns for outlier analysis
selected_columns = [
    'Population',
    'Per capita GNI',
    'Gross Domestic Product (GDP)',
    'Gross National Income(GNI) in USD',
    'Exports of goods and services',
    'Imports of goods and services',
    'AMA exchange rate',
    'IMF based exchange rate',
    'Total Value Added'
]

# Creating a DataFrame with selected columns
selected_df = global_economy_df[selected_columns]

# Calculate z-scores for each column
z_scores = (selected_df - selected_df.mean()) / selected_df.std()

# Set the threshold for outliers
threshold = 3  # Threshold is based on the empirical rule (68-95-99.7 rule) for normally distributed data

# Identify outliers
outliers_selected = (z_scores > threshold) | (z_scores < -threshold)

# Display outlier count for each column
outlier_count_by_column = outliers_selected.apply(np.sum)
print("Outlier Count by Column:")
print(outlier_count_by_column)
```

**Before we handle the outliers, let's visualise the outliers for calrity**

```python
# Import Library
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

# Generate box plots for each selected column, displaying two columns per row
for i, column in enumerate(outliers_selected.columns, 1):
    plt.subplot(5, 2, i)
    sns.boxplot(x=global_economy_df[column], color='skyblue')
    plt.axvline(global_economy_df[column].min(), color='r', linestyle='--', label='Min Value')
    plt.axvline(global_economy_df[column].max(), color='g', linestyle='--', label='Max Value')
    plt.title(column)
    plt.legend()

    # Annotate max value
    max_value = global_economy_df[column].max()
    plt.annotate(f'Max: {max_value:.2f}', xy=(0.95, 0.9), xycoords='axes fraction',
                 ha='right', va='center', color='red', fontsize=8)

    # Annotate min value
    min_value = global_economy_df[column].min()
    plt.annotate(f'Min: {min_value:.2f}', xy=(0.95, 0.8), xycoords='axes fraction',
                 ha='right', va='center', color='green', fontsize=8)

# Adjust layout
plt.tight_layout()
plt.show()
```

Outliers can significantly impact statistical analyses and machine learning models by introducing bias into the results. To address this, the analysis begins by narrowing down specific columns relevant to the analytical question. A new DataFrame, selected_df, is created to contain the selected columns. Using the NumPy library for mathematical operations and array manipulations, z-scores are calculated for each column by subtracting the mean and dividing by the standard deviation. Outliers are identified using the empirical rule (68-95-99.7) for normally distributed data, with a threshold of 3. The absolute z-scores are compared against this threshold to flag outliers. Counts of outliers in each column are displayed, providing insight into the prevalence of extreme values. To enhance understanding, box plots are generated for each column, with annotations highlighting maximum and minimum values.

Subsequently, the percentage of outliers in each column is calculated and reported. The results show that outlier percentages range from approximately 0.93% to 1.91%, well below the commonly accepted threshold of 5%. Given the low percentage and the context of the analysis, the decision is made not to remove these outliers, as they are unlikely to significantly affect the results. Furthermore, the dataset's nature, which includes countries with both exceptionally high and low economic figures over time, aligns with domain knowledge, recognizing these deviations as inherent characteristics of the data.

**K-means Clustering - we applying an unsupervised machine learning algorithm to group countries into clusters based on their economic features. And then employ scatter plot to visually represents these clusters in a 2D space.**

```python
# Import Library
from sklearn.cluster import KMeans
import plotly.express as px

# Select numerical features for clustering
cluster_features = ['Population', 'Gross Domestic Product (GDP)', 'Per capita GNI', 'Exports of goods and services', 'Imports of goods and services', 'Year']

# Filter data for clustering features
cluster_data = global_economy_df_cleaned[cluster_features]

# Standardise the features (important for K-means)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)

# Explicitly set n_init to suppress the warning - (I was getting some warnings about `n_init` )
num_clusters = 10
kmeans = KMeans(n_clusters=num_clusters, n_init=10, random_state=42)
global_economy_df_cleaned['Cluster'] = kmeans.fit_predict(cluster_data_scaled)

# Print the cluster assignments
print("K-means Cluster Assignments:")
print(global_economy_df_cleaned[['Population', 'Gross Domestic Product (GDP)', 'Per capita GNI', 'Exports of goods and services', 'Imports of goods and services', 'Year', 'Cluster']])
```

```python
# Visualise the clusters using a scatter plot
import plotly.express as px


# Scatter plot for visualising clusters
fig = px.scatter(
    global_economy_df_cleaned,
    x='Per capita GNI',
    y='Gross Domestic Product (GDP)',
    color='Cluster',
    size='Population',
    hover_data=['Country'],
    title='K-means Clustering of Countries based on Economic Features',
    labels={'Per capita GNI': 'Per capita GNI', 'Gross Domestic Product (GDP)': 'GDP'},
    template='plotly_dark',
)

# Show the figure
fig.show()

```

**SUPERVISED MACHINE LEARNING ALGORITHM USING LINEAR REGRESSION - We will apply this to predict the Gross Domestic Product (GDP) based on other economic features. Linear regression will assume a linear relationship between the input features and the target variable and then try to find the best-fitting line that represents this relationship.**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
import numpy as np
```

**Lets Select features and target variable we will use for this process**

```python
# Select features and target variable
features = ['Population', 'Per capita GNI', 'Exports of goods and services', 'Imports of goods and services', 'Year']
target_variable = 'Gross Domestic Product (GDP)'

# Split data into features and target variable
X = global_economy_df_cleaned[features]
y = global_economy_df_cleaned[target_variable]
```

**To avoid unintended errors, it is important to standardise or normalise the numerical values of the features and target variable.The original values of the target variable 'Gross Domestic Product (GDP)' might have different magnitudes from the features and as such, we have to scale them to ensures that all values are within a similar range. This is particularly important for algorithms that are sensitive to the scale of the input features, such as linear regression.**

```python
# Scale the target variable (GDP) using Min-Max Scaling
scaler_gdp = MinMaxScaler()
y_scaled = scaler_gdp.fit_transform(y.values.reshape(-1, 1))
```

**Initialise and train the linear regression model**

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_scaled, test_size=0.2, random_state=42)

# Initialize the linear regression model
model = LinearRegression()

# Train the model on the training set
model.fit(X_train, y_train)
```

**Make prediction**

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

**Inversely transform the scaled predictions to get the original scale - Here is why this is important: This helps to convert the predicted GDP values from the scaled form back to their original scale. This step is crucial for interpretation, comparison with actual values, and presenting the model results in a format that aligns with the original data units.**

```python
# Inverse transform the scaled predictions to get the original scale
y_pred_original_scale = scaler_gdp.inverse_transform(y_pred)
```

**Evaluate Model**

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Print the coefficients of the model
print("Coefficients:")
for feature, coefficient in zip(features, model.coef_[0]):
    print(f"{feature}: {coefficient}")

# Print the intercept of the model
print(f"Intercept: {model.intercept_[0]}")
```

**Visualise model**
